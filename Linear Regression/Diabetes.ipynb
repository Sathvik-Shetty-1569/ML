```python
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
def calculate_VIF(dataframe):
    vif = pd.DataFrame()
    vif['Feature'] = dataframe.columns
    
    vif['VIF'] = [variance_inflation_factor(dataframe.values,i) for i in range(dataframe.shape[1])]
    
    return vif.sort_values('VIF',ascending=False).reset_index(drop=True)
    
```

```python
vif = calculate_VIF(X_train)
vif
```

```python
scalar = StandardScaler()
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
X_train_scaled = np.array(X_train_scaled)
X_test_scaled = np.array(X_test_scaled)
y_train_arr = np.array(y_train)
y_test_arr = np.array(y_test)
```

```python
def linear_regression_gradient(X,Y,learning_rate,n_iterations):
    n_rows = X.shape[0]
    n_columns = X.shape[1]
    w = np.zeros(n_columns)
    b = 0.0
    loss_history = []
    
    for i in range(n_iterations):
        y_pred = X @ w + b
    
        error = y_pred - Y
        
        loss = np.mean(error**2)
        
        loss_history.append(loss)
        
        dw = (2/n_rows)*(X.T @ error)
        db = (2/n_rows)*np.sum(error)
        
        w = w - learning_rate * dw
        b = b - learning_rate * db
        
        if i % 100 == 0:
            print(f"Iteration {i:4d} | Loss (MSE): {loss:.4f}")
 
    return w, b, loss_history
    
    
```

```python
print("Training from-scratch Linear Regression with Gradient Descent...\n")
w_scratch, b_scratch, loss_history = linear_regression_gradient(
    X_train_scaled, y_train_arr, learning_rate=0.2, n_iterations=1000
)
 
print("\nFinal weights (scratch):", w_scratch)
print("Final bias (scratch):", b_scratch)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(loss_history)
plt.xlabel("Iteration")
plt.ylabel("Mean Squared Error (Loss)")
plt.title("Gradient Descent Convergence - Loss vs Iteration")
plt.grid(True)
plt.show()
 
print("\nLoss curve saved.")
print("If this curve flattens out, gradient descent has converged.")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train_arr)
print("Comparison of weights:\n")
comparison_df = pd.DataFrame({
    "Feature": X.columns,
    "Scratch Weights (GD)": w_scratch,
    "Sklearn Weights (OLS)": model.coef_
})
print(comparison_df)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
def adjusted_r2(r2, n_samples, n_features):
    """
    Adjusted R² formula:
    Adj_R2 = 1 - (1 - R2) * (n - 1) / (n - p - 1)
    where n = number of samples, p = number of features
 
    WHY this formula: as p (features) increases, the denominator
    (n - p - 1) shrinks, which makes the penalty term larger,
    pulling Adjusted R2 DOWN unless the new feature genuinely
    improves R2 enough to offset it.
    """
    return 1 - (1 - r2) * (n_samples - 1) / (n_samples - n_features - 1)
 
 
def evaluate_model(model_name, y_true_train, y_pred_train, y_true_test, y_pred_test, n_features):
    """Print a clean comparison of R2, Adjusted R2, and RMSE for train vs test."""
    r2_train = r2_score(y_true_train, y_pred_train)
    r2_test = r2_score(y_true_test, y_pred_test)
 
    adj_r2_train = adjusted_r2(r2_train, len(y_true_train), n_features)
    adj_r2_test = adjusted_r2(r2_test, len(y_true_test), n_features)
 
    rmse_train = np.sqrt(mean_squared_error(y_true_train, y_pred_train))
    rmse_test = np.sqrt(mean_squared_error(y_true_test, y_pred_test))
 
    print(f"\n--- {model_name} ---")
    print(f"{'Metric':<15}{'Train':>10}{'Test':>10}")
    print(f"{'R2':<15}{r2_train:>10.4f}{r2_test:>10.4f}")
    print(f"{'Adjusted R2':<15}{adj_r2_train:>10.4f}{adj_r2_test:>10.4f}")
    print(f"{'RMSE':<15}{rmse_train:>10.4f}{rmse_test:>10.4f}")
 
    return {
        "model": model_name,
        "r2_train": r2_train, "r2_test": r2_test,
        "adj_r2_train": adj_r2_train, "adj_r2_test": adj_r2_test,
        "rmse_train": rmse_train, "rmse_test": rmse_test
    }
 
 
# %%
# Predictions from our scratch model
y_train_pred_scratch = X_train_scaled @ w_scratch + b_scratch
y_test_pred_scratch = X_test_scaled @ w_scratch + b_scratch
 
# Predictions from sklearn model
y_train_pred_sklearn = model.predict(X_train_scaled)
y_test_pred_sklearn = model.predict(X_test_scaled)
 
n_features = X_train_scaled.shape[1]
 
results_scratch = evaluate_model(
    "Scratch Linear Regression (Gradient Descent)",
    y_train_arr, y_train_pred_scratch, y_test_arr, y_test_pred_scratch, n_features
)
 
results_sklearn = evaluate_model(
    "Sklearn Linear Regression (OLS)",
    y_train_arr, y_train_pred_sklearn, y_test_arr, y_test_pred_sklearn, n_features
)
```

```python
from scipy import stats as scipy_stats
 
# We'll analyze residuals from the sklearn model on the TEST set
# (test set residuals are what actually matter - they reflect
# real generalization behavior, not just memorized training fit)
residuals_test = y_test_arr - y_test_pred_sklearn
 
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
 
# --- Plot 1: Residuals vs Predicted ---
axes[0].scatter(y_test_pred_sklearn, residuals_test, alpha=0.6, edgecolor="k")
axes[0].axhline(0, color="red", linestyle="--", linewidth=1.5)
axes[0].set_xlabel("Predicted Values")
axes[0].set_ylabel("Residuals (Actual - Predicted)")
axes[0].set_title("Residuals vs Predicted")
 
# --- Plot 2: Q-Q Plot ---
scipy_stats.probplot(residuals_test, dist="norm", plot=axes[1])
axes[1].set_title("Q-Q Plot of Residuals")
 
# --- Plot 3: Histogram of Residuals ---
axes[2].hist(residuals_test, bins=20, edgecolor="k", alpha=0.7)
axes[2].axvline(0, color="red", linestyle="--", linewidth=1.5)
axes[2].set_xlabel("Residual Value")
axes[2].set_ylabel("Frequency")
axes[2].set_title("Histogram of Residuals")
 
plt.tight_layout()
plt.show()
 
print("Residual diagnostics saved.")
print("\nWhat to check:")
print("1. Residuals vs Predicted - is there a funnel shape (heteroscedasticity)")
print("   or a curve (non-linearity)? Or does it look like a random cloud?")
print("2. Q-Q Plot - do points follow the diagonal line (normal residuals)")
print("   or do they curve away at the ends (heavy tails)?")
print("3. Histogram - roughly bell-shaped and centered at 0?")
print("\nIf residuals show a clear pattern, R2 alone would have HIDDEN this")
print("problem from you - this is why experts always plot residuals.")
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
alpha = 1.0  # regularization strength - higher = more shrinkage
 
ridge = Ridge(alpha=alpha)
ridge.fit(X_train_scaled, y_train_arr)
 
lasso = Lasso(alpha=alpha)
lasso.fit(X_train_scaled, y_train_arr)
 
# Compare coefficients side by side
coef_comparison = pd.DataFrame({
    "Feature": X.columns,
    "Linear (no penalty)": model.coef_,
    "Ridge (L2)": ridge.coef_,
    "Lasso (L1)": lasso.coef_
})
print("Coefficient comparison:\n")
print(coef_comparison)
 
print("\nNumber of features Lasso zeroed out:", np.sum(lasso.coef_ == 0))
print("Number of features Ridge zeroed out:", np.sum(ridge.coef_ == 0))
 
# Evaluate both
results_ridge = evaluate_model("Ridge Regression", y_train_arr,
                                ridge.predict(X_train_scaled), y_test_arr,
                                ridge.predict(X_test_scaled), n_features)
 
results_lasso = evaluate_model("Lasso Regression", y_train_arr,
                                lasso.predict(X_train_scaled), y_test_arr,
                                lasso.predict(X_test_scaled), n_features)
```

```python
alphas = np.logspace(-3, 2, 50)  # 50 values from 0.001 to 100, evenly spaced in log scale
lasso_coefs = []
 
for a in alphas:
    lasso_temp = Lasso(alpha=a, max_iter=10000)
    lasso_temp.fit(X_train_scaled, y_train_arr)
    lasso_coefs.append(lasso_temp.coef_)
 
lasso_coefs = np.array(lasso_coefs)  # shape: (n_alphas, n_features)
 
plt.figure(figsize=(10, 6))
for i, feature_name in enumerate(X.columns):
    plt.plot(alphas, lasso_coefs[:, i], label=feature_name)
 
plt.xscale("log")  # log scale because alpha spans several orders of magnitude
plt.xlabel("Alpha (Regularization Strength) - log scale")
plt.ylabel("Coefficient Value")
plt.title("Lasso Coefficient Shrinkage as Alpha Increases")
plt.legend(loc="best", fontsize=8)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.grid(True, alpha=0.3)
plt.show()
 
print("Lasso shrinkage plot saved.")
print("Notice how, as alpha increases (moving right), each line eventually")
print("hits exactly 0 and STAYS there - this is Lasso doing feature selection.")
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=3, include_bias=False)
 
X_train_poly = poly.fit_transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)
 
print(f"Original number of features: {X_train_scaled.shape[1]}")
print(f"Number of features after degree-3 polynomial expansion: {X_train_poly.shape[1]}")
 
# Train plain linear regression on this huge feature set - NO regularization
overfit_model = LinearRegression()
overfit_model.fit(X_train_poly, y_train_arr)
 
results_overfit = evaluate_model(
    "Overfit Model (Degree-3 Polynomial, No Regularization)",
    y_train_arr, overfit_model.predict(X_train_poly),
    y_test_arr, overfit_model.predict(X_test_poly),
    X_train_poly.shape[1]
)
 
print("\n>>> COMPARE this to the original simple Linear Regression results above. <<<")
print(">>> A big TRAIN vs TEST gap here = OVERFITTING. <<<")
```

```python
ridge_poly = Ridge(alpha=10.0)  # higher alpha = stronger correction
ridge_poly.fit(X_train_poly, y_train_arr)
 
results_ridge_poly = evaluate_model(
    "Ridge on Polynomial Features (Fixing Overfitting)",
    y_train_arr, ridge_poly.predict(X_train_poly),
    y_test_arr, ridge_poly.predict(X_test_poly),
    X_train_poly.shape[1]
)
 
print("\n>>> Compare train/test gap here vs the overfit model above. <<<")
print(">>> Ridge should shrink the gap by penalizing huge coefficients. <<<")
 
 
# %% [markdown]
# ## STEP 13: Final Summary Table - Everything Side by Side
 
# %%
all_results = pd.DataFrame([
    results_scratch, results_sklearn, results_ridge,
    results_lasso, results_overfit, results_ridge_poly
])
 
# Show only the most important columns, rounded for readability
summary = all_results[["model", "r2_train", "r2_test", "adj_r2_train", "adj_r2_test",
                        "rmse_train", "rmse_test"]].round(4)
print(summary.to_string(index=False))
```
